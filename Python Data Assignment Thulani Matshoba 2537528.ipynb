```python

# Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# This line loads data from GrowLocations.cvs into a dataframe called grow_locations
grow_locations = pd.read_csv("GrowLocations.csv")

# This line filters data based on latitude and longitude using only data the lay between the ranges 
# of Latitude 50.681 - 57.985 and Locations -10.595 to 1.6848 
filter_conditions = (grow_locations['Latitude'].between(50.681, 57.985)) & \
                    (grow_locations['Longitude'].between(-10.592, 1.6848))
filtered_grow_locations = grow_locations[filter_conditions]

# This line loads the UK map7.png which also defines the longitude and latitude boundaries 
uk_map_image = plt.imread("map7.png")
map_extent = [-10.592, 1.6848, 50.681, 57.985]

# This line Creates a plot of size 10x10
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(uk_map_image, extent=map_extent)

# This line plot each sensor location. The loop iterates through each row in the dataframe
for _, location in filtered_grow_locations.iterrows():
    ax.scatter(location['Longitude'], location['Latitude'], color='red', marker='o', s=5)

# This line sets the tiltle and lables the x-axis and y-axis 
ax.set_title('Sensor Locations')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display the plot
plt.show()
```
